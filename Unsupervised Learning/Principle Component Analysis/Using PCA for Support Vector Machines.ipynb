```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

Below is scaled data of 30 dimensions

```python
scaled_data[0]
```

```python
x_pca[0]
```

Above is PCAfied data i.e 30 dimensions got converted to 2 using pca.fit_transform 

```python
sddf = pd.DataFrame(x_pca)
```

```python
sddf['malignant'] = cancer.target
```

```python
sddf.drop('class',axis=1,inplace=True)
```

```python
sddf
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'])
plt.xlabel('First Principle Component')
plt.ylabel('Second Principe Component')
```

```python
plt.scatter(pca.components_[0],pca.components_[1])
```

```python
df_comp = pd.DataFrame(pca.components_,columns= cancer['feature_names'])
```

```python
df_comp
```

```python
sns.heatmap(df_comp,cmap='plasma')
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel = 'linear')
```

```python
from sklearn.model_selection import train_test_split
```

```python
sddf[0]
```

```python
X_train,X_test,y_train,y_test = train_test_split(sddf[[0,1]],sddf['malignant'],test_size=0.6,random_state=101)
```

```python
svc.fit(X_train,y_train)
```

```python
pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,pred))
```

```python
print(confusion_matrix(y_test,pred))
```
